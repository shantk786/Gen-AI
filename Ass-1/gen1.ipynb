```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Training data
data = np.array([55, 60, 65, 70, 72, 75, 80, 85])

# Step 2: Estimate parameters
mean = np.mean(data)
std = np.std(data)

print("Estimated Mean:", mean)
print("Estimated Std Dev:", std)

# Step 3: Generate new samples
generated_data = np.random.normal(mean, std, 1000)

# Visualization
plt.hist(generated_data, bins=30, density=True)
plt.title("Generated Data Distribution")
plt.xlabel("Marks")
plt.ylabel("Probability Density")
plt.show()
```
