```python
import torch
from diffusers import StableDiffusionPipeline
import cv2
from PIL import Image
import numpy as np
```

```python
# Load model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id).to("cuda")
```

```python
prompt = "a bright fantasy world with floating mountains, glowing rivers, crystal forests, magical cities, cinematic lighting"
```

```python
frames = []

# Generate frames
for i in range(10):
    image = pipe(prompt).images[0]
    frame = np.array(image)
    frames.append(frame)
```

```python
# Create video
height, width, layers = frames[0].shape
video = cv2.VideoWriter("fantasy_animation.mp4",
                        cv2.VideoWriter_fourcc(*'mp4v'),
                        2,
                        (width, height))
```

```python
for frame in frames:
    video.write(frame)

video.release()

print("Video saved as fantasy_animation.mp4")
```
